#### 1. Даны значения зарплат из выборки выпускников:
100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
import numpy as np
```

```python
sal = np.array(sorted([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]), order='F')
sal
```

Формула среднеквадратичного отклонения:

$$\sigma = \sqrt{\frac{\sum\limits_{i=1}^{n} (x_i - \overline{x})^2}{n}}$$

```python
def params(arr):
    
    n = len(arr)
    mean_ = sum(arr) / n
    sumvar = sum([(x - mean_) ** 2 for x in arr])

    DXzero = sumvar / n
    std_zero = DXzero ** 0.5
    
    DXone = sumvar / (n - 1)
    std_one = DXone ** 0.5
    
    return mean_, std_zero, std_one, DXzero, DXone
```

```python
meansal, sigma0, sigma1, DX0, DX1 = params(sal)
print(f'Среднее арифметическое = {meansal}\n')
print(f'Смещенная дисперсия = {DX0}')
print(f'Cреднеквадратичное отклонение на основании смещённой оценки дисперсии = {sigma0}\n')
print(f'Несмещенная дисперсия = {DX1}')
print(f'Cреднеквадратичное отклонение на основании несмещённой оценки дисперсии = {sigma1}')
```

```python
print(f'Проверка:')
print(f'среднее арифметическое  = {sal.mean()}\n')
print(f'Смещенная дисперсия = {sal.var(ddof=0)}')
print(f'Cреднеквадратичное отклонение на основании смещённой оценки дисперсии = {sal.std(ddof=0)}\n')
print(f'Несмещенная дисперсия = {sal.var(ddof=1)}')
print(f'Cреднеквадратичное отклонение на основании несмещённой оценки дисперсии = {sal.std(ddof=1)}')
```

#### 2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых.
Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

Событие наступает в нескольких случаях:

    1. Первый ящик - 0 белых , второй - 3
    
    2. Первый ящик - 1 белый , второй - 2

    3. Первый ящик - 2 белых , второй - 1

Т.е., вероятность наступления события равна сумме вероятностей перечисленных случаев (ИЛИ). А каждый из случаев наступает при одновременном наступлении событий в каждом ящике (И)

1 вариант, первый ящик, вероятность = количеству сочетаний (все небелые, т.е. 2 из 3х имеющихся) деленному на количество всех сочетаний небелых в первом ящике:

$$C_3^2 = \frac{3!}{2!(3 - 2)!}$$

т.е. 3. 

$$C_8^2 = \frac{8!}{2!(8 - 2)!}$$

8 * 7 / 2 = 28

Итого вероятность того, что из первого ящика вынут 0 белых = 3/28

1 вариант, второй ящик (3 из 5 белых, делить на 4 из 12):

$$C_5^3 = \frac{5!}{3!(5 - 3)!}$$

5 * 4 / 2 = 10

$$C_12^4 = \frac{12!}{4!(12 - 4)!}$$

12 * 11 * 10 * 9 / (4 * 3 * 2) = 11 * 5 * 9 = 495

Итого вероятность того, что из второго ящика вынут 3 белых = 10/495 = 2 / 99

Вероятность того, что из 1 ящика вынут 0 белых, а из второго - 3 = 3/28 * 2/99 = 1 / (14 * 33) = 1/462 или 0.0021645

Второй вариант, первый ящик (1 белый):

$$C_5^1 = \frac{5!}{1!(5 - 1)!}$$

Итого 5, Вероятность 5/28

Второй ящик (2 белых из 5ти) = 10 / 495 = 2 / 99

Вероятность второго варианта = 5/28 * 2/ 99 = 5/1386 = 0.0036075

Третий вариант, первый ящик (2 белых из 5) = 10/28 = 5/14

Второй ящик (1 белый из 5) = 5 / 495 = 1/99

Вероятность 3го варианта = 5/14 * 1/99 = 5/1386 = 0.0036075

#### Ответ:  Вероятность всех трех вариантов = 1/462 + 5/1386 + 5/1386 = 13/1386 = 0.0093795

#### 3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень.
Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

Прежде чем узнали результат выстрела, мы располагали только априорной вероятностью того, что выстрел был произведен, и она равнялась 1/3 — если учесть, что у всех спортсменов были равные возможности выстрелить.

Зная результат выстрела, мы можем заново оценить вероятности для каждого спортсмена — ведь точность у них разная. Эти вероятности будут апостериорными, так как они получены после поступления информации о том, что выстрел был удачным.

Событием $A$ будем считать попадание в мишень, а событиями $B_1, B_2$ и $B_3$ — что выстрел совершил первый, второй или третий спортсмен:

$$P(A\:|\:B_1) = 0.9$$

$$P(A\:|\:B_2) = 0.8$$

$$P(A\:|\:B_3) = 0.6$$

Вероятность того, что стрелял первый спортсмен — при условии, что выстрел был удачным, — будем находить по формуле:

$$P(B_1\:|\:A) = \frac{P(B_1) \cdot P(A\:|\:B_1)}{P(A)}$$

Все величины подставим из условия, а знаменатель дроби распишем по формуле полной вероятности:

$$P(B_1\:|\:A) = \frac{\frac{1}{3} \cdot 0.9}{\frac{1}{3} \cdot 0.9 + \frac{1}{3} \cdot 0.8 + \frac{1}{3} \cdot 0.6} = \frac{9}{23}$$

Аналогично, вероятность для второго спортсмена = 8/23, для третьего = 6/23

#### Ответ: Вероятность того, что выстрел произведен: a). первым спортсменом = 9/23 б). вторым спортсменом = 8/23 в). третьим спортсменом = 6/23.

#### 4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе.
Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

Прежде чем узнали результат сдачи, мы располагали только априорной вероятностью того, что студены сдавали сессию, и она равнялась 1/4 для факульетов А и В, и 1/2 для факультета С — если учесть, что у всех студентов были равные возможности для сдачи сессии.

Зная результат сдачи, мы можем заново оценить вероятности для каждого факульета — ведь "успеваемость" у них разная. Эти вероятности будут апостериорными, так как они получены после поступления информации о том, что сессия была сдана.

Событием $A$ будем считать успешную сдачу, а событиями $B_1, B_2$ и $B_3$ — что сдал студент соответсвенно факультета А,В и С:

$$P(A\:|\:B_1) = 0.8$$

$$P(A\:|\:B_2) = 0.7$$

$$P(A\:|\:B_3) = 0.9$$

Вероятность того, что сдававший студент учится на фак.А — при условии, что сессия была сдана, — будем находить по формуле:

$$P(B_1\:|\:A) = \frac{P(B_1) \cdot P(A\:|\:B_1)}{P(A)}$$

Все величины подставим из условия, а знаменатель дроби распишем по формуле полной вероятности:

$$P(B_1\:|\:A) = \frac{\frac{1}{4} \cdot 0.8}{\frac{1}{4} \cdot 0.8 + \frac{1}{4} \cdot 0.7 + \frac{1}{2} \cdot 0.9} = \frac{8}{33}$$

Аналогично, вероятность для второго спортсмена = 7/33, для третьего = 18/33

#### Ответ: Вероятность того, что студент учится: a). на факультете A = 8/33 б). на факультете B = 7/33 в). на факультете C = 18/33.
